```python
rating_popular_books=rating_popular_books.drop_duplicates('Title')
```

```python
piv=rating_popular_books.pivot(index='Title',columns='User_id',values='rating').fillna(0)
piv.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
rating_matrix=csr_matrix(piv.values)
```

```python
print(rating_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors

knn=NearestNeighbors(metric='cosine')
knn.fit(rating_matrix)                    
```

```python
query_index=np.random.choice(piv.shape[0])
distance,indices=knn.kneighbors(piv.iloc[query_index, :].values.reshape(1, -1),n_neighbors=5)

for i in range(0,len(distance.flatten())):
    if i==0:
        print('Recommend for : {0} '.format(piv.index[query_index]))
    else:
        print('{0}: {1}, with distances of {2}: '.format(i,piv.index[indices.flatten()[i]],distance.flatten()[i]))
```

```python
from collections import Counter
import math
A=Counter(list('abcd'))
B=Counter(list('ab'))

print(A)
print(B)


terms=set(A).union(B)
print(terms)

dotprod=sum(A.get(K,0) * B.get(K,0) for K in terms)
# print(dotprod)

magA=math.sqrt(sum(A.get(i,0)**2 for i in terms))
magB=math.sqrt(sum(B.get(i,0)**2 for i in terms))

print('Cosine Similarity: ',dotprod/magA*magB)
```
