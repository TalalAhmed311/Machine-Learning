```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
slope=[]
intercept=[]
error=[]   
def gradient_descent(x,y):
    iterations=1000
    n=len(x)
    m=b=0
    learning_rate=0.0001
    for i in range(iterations):
        y_pred=m*x+b
        
        cost=(1/n)*sum(v**2 for v in (y-y_pred))
        error.append(cost)
        m_deriv=-(2/n)*sum(x*(y-y_pred))
        b_deriv=-(2/n)*sum(y-y_pred)
        m=m-learning_rate*m_deriv
        b=b-learning_rate*b_deriv
        
        slope.append(m)
        intercept.append(b)
        print('m {}, b {}, cost {}, iteration {} '.format(m,b,cost,i))
        
```

```python
gradient_descent(x,y)

# plt.plot(m,b)
# plt.show()
```
