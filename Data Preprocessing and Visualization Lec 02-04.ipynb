```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set(color_codes=True)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 10000),
    'x2': np.random.normal(5, 3, 10000),
    'x3': np.random.normal(-5, 5, 10000)
})
```

```python
# plotting data

df.plot.kde()
```

```python

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
data_tf = ss.fit_transform(df)
data_tf
```

```python

df = pd.DataFrame(data_tf, columns=['x1','x2','x3'])
```

```python

df.plot.kde()
```

```python
df = pd.DataFrame({
    # positive skew
    'x1': np.random.chisquare(8, 1000),
    # negative skew 
    'x2': np.random.beta(8, 2, 1000) * 40,
    # no skew
    'x3': np.random.normal(50, 3, 1000)
})
df
```

```python

df.plot.kde()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

mm = MinMaxScaler()
data_tf = mm.fit_transform(df)
data_tf
```

```python
df = pd.DataFrame(data_tf,columns=['x1','x2','x3'])
df.plot.kde()
```

```python
#Robust Scaler
df = pd.DataFrame({
    # Distribution with lower outliers
    'x1': np.concatenate([np.random.normal(20, 1, 1000), np.random.normal(1, 1, 25)]),
    # Distribution with higher outliers
    'x2': np.concatenate([np.random.normal(30, 1, 1000), np.random.normal(50, 1, 25)]),
})
df[df['x1']>3]
```

```python
df.plot.kde()
```

```python

from sklearn.preprocessing import RobustScaler
robustscaler = RobustScaler()
```

```python

data_tf = robustscaler.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1','x2'])
df
```

```python

df.plot.kde()
```

```python
sentences=[
    'How to learn data Science',
    'How to Machine learning from Sratch',
    'Building a Movie recomendation System using Python',
    'I am intrested in Natural Language Processing in Machine learning'
]
df=pd.DataFrame({'Text':sentences})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
cv.fit_transform(df.Text).toarray()
```

```python
cv.transform(df.Text).toarray()
```

```python
cv.vocabulary_
```

```python
cv.get_feature_names()
```

```python
cv=CountVectorizer(ngram_range=(2,2))
cv.fit_transform(df.Text).toarray()
```

```python
cv.vocabulary_
```

```python
cv.get_feature_names()
```

```python
cv=CountVectorizer(vocabulary=['machine learning'],ngram_range=[2,2])
cv.fit_transform(df.Text).toarray()
```

```python
#Seaborn

#Making a relationship plot
#Flights DataSet 
#Col: Year, Months, Passenger

flights=sns.load_dataset('flights')
# flights

sns.relplot(x='passengers',y='month',data=flights)
```

```python
#Relationship in tips dataset
#Col:total_bill,tip,sex,smoker,day,time,size

tips=sns.load_dataset('tips')
sns.relplot(x='total_bill',y='tip',data=tips,kind='line')
```

```python
#Categorical Plot

sns.catplot(x='sex',y='tip',data=tips)
```
